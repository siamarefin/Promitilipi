```python
import cv2
import os
import matplotlib.pyplot as plt

# Paths
images_dir = '/kaggle/input/promitilipi/PromitoLipi2/PromitoLipi2/WordImages(bmp)/'
annotations_dir = '/kaggle/input/promitilipi/yolo_annotations/'

# Function to visualize the image with bounding boxes
def visualize_image_with_yolo_annotations(image_path, annotation_path):
    # Load the image using OpenCV (BGR format)
    image = cv2.imread(image_path)
    
    # Convert the image from BGR to RGB for matplotlib compatibility
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    # Read YOLO annotations from the text file
    with open(annotation_path, 'r') as file:
        annotations = file.readlines()
    
    # Loop through each annotation
    for annotation in annotations:
        # Extract the values (class_id, x_center, y_center, width, height)
        parts = annotation.strip().split()
        class_id, x_center, y_center, width, height = map(float, parts)
        
        # Convert the normalized values to absolute pixel values
        img_width, img_height = image.shape[1], image.shape[0]
        x_center = int(x_center * img_width)
        y_center = int(y_center * img_height)
        box_width = int(width * img_width)
        box_height = int(height * img_height)
        
        # Calculate the top-left corner and bottom-right corner of the bounding box
        xmin = int(x_center - box_width / 2)
        ymin = int(y_center - box_height / 2)
        xmax = int(x_center + box_width / 2)
        ymax = int(y_center + box_height / 2)
        
        # Draw the bounding box (on the RGB image for display)
        cv2.rectangle(image_rgb, (xmin, ymin), (xmax, ymax), (0, 255, 0), 2)
        cv2.putText(image_rgb, str(int(class_id)), (xmin, ymin-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)
    
    # Display the image with matplotlib (in the notebook)
    plt.figure(figsize=(10, 6))
    plt.imshow(image_rgb)
    plt.axis('off')  # Hide axis labels
    plt.show()

# Counter to display only the first 10 images
counter = 0
max_images = 10

# Loop through the images and their corresponding annotation files
for image_filename in os.listdir(images_dir):
    if image_filename.endswith('.bmp') and counter < max_images:
        # Construct the full image path
        image_path = os.path.join(images_dir, image_filename)
        
        # Construct the corresponding annotation path
        annotation_filename = image_filename.replace('.bmp', '.txt')
        annotation_path = os.path.join(annotations_dir, annotation_filename)
        
        if os.path.exists(annotation_path):
            # Visualize the image with annotations
            visualize_image_with_yolo_annotations(image_path, annotation_path)
        
        # Increment the counter
        counter += 1
```

```python
from ultralytics import YOLO
import cv2

# Load the trained YOLOv8 model
model = YOLO('best_yolov8_model.pt')  # Load your trained model

# Define the path to the test image and its annotation file
image_path = '/kaggle/input/promitilipi/split_dataset/test/images/N3.bmp'
annotation_path = '/kaggle/input/promitilipi/split_dataset/test/labels/N3.txt'

# Define the class names (Bangla characters from 0 to 92)
class_names = [
    'blank', 'অ', 'ই', 'ঈ', 'উ', 'ঊ', 'ঋ', 'এ', 'ঐ', 'ও', 'ঔ', 'ক', 'খ', 'গ', 'ঘ', 'ঙ', 'চ', 'ছ', 'জ', 'ঝ',
    'ঞ', 'ট', 'ঠ', 'ড', 'ঢ', 'ণ', 'ত', 'থ', 'দ', 'ধ', 'ন', 'প', 'ফ', 'ব', 'ভ', 'ম', 'য', 'র', 'ল', 'শ', 'ষ',
    'স', 'হ', 'ড়', 'ঢ়', 'য়', 'ৎ', 'ঃ', 'ং', 'ঁ', '০', '১', '২', '৩', '৪', '৫', '৬', '৭', '৮', '৯', 'া', 'ি', 'ী',
    'ে', 'ু', 'faka', '্র', '্য', 'ক্ষ', 'ন্ত', 'ত্র', 'ঙ্গ', 'স্থ', 'স্ব', 'ক্ত', 'স্ত', 'ন্দ', 'চ্ছ', 'দ্ধ',
    'ন্ত্র', 'ফাকা', 'ত্ত', 'ষ্ট', 'ন্ন', 'ল্প', 'ম্প', 'faka', 'ূ', 'ৃ', 'ৈ', 'faka', 'ৌ', '।'
]

# Load the ground truth class from the annotation file (handle multiple classes)
ground_truth_classes = []
with open(annotation_path, 'r') as f:
    for line in f.readlines():
        parts = line.strip().split()
        class_id = int(parts[0])  # Ground truth class ID
        ground_truth_classes.append(class_id)

# Make prediction on the image
results = model.predict(image_path, imgsz=640)  # Adjust imgsz if needed

# Extract predicted class IDs from the results
predicted_classes = results[0].boxes.cls.tolist()  # Get predicted class IDs

# Calculate accuracy for this single prediction (correct if predicted class matches ground truth)
correct_predictions = 0
for pred_class in predicted_classes:
    if int(pred_class) in ground_truth_classes:
        correct_predictions += 1

# Calculate accuracy
accuracy = (correct_predictions / len(ground_truth_classes)) * 100

# Get the predicted class names for visualization
predicted_class_names = [class_names[int(cls)] for cls in predicted_classes]
ground_truth_class_names = [class_names[cls] for cls in ground_truth_classes]

# Print the predicted classes, ground truth, and accuracy
print(f"Predicted Classes: {predicted_class_names}")
print(f"Ground Truth Classes: {ground_truth_class_names}")
print(f"Accuracy: {accuracy:.2f}%")
```

```python
from ultralytics import YOLO
import cv2
import os
import random

# Load the trained YOLOv8 model
model = YOLO('best_yolov8_model.pt')  # Load your trained model

# Define the path to the test images and their annotations
test_images_dir = '/kaggle/input/promitilipi/split_dataset/test/images/'
test_annotations_dir = '/kaggle/input/promitilipi/split_dataset/test/labels/'

# Define the class names (Bangla characters from 0 to 92)
class_names = [
    'blank', 'অ', 'ই', 'ঈ', 'উ', 'ঊ', 'ঋ', 'এ', 'ঐ', 'ও', 'ঔ', 'ক', 'খ', 'গ', 'ঘ', 'ঙ', 'চ', 'ছ', 'জ', 'ঝ',
    'ঞ', 'ট', 'ঠ', 'ড', 'ঢ', 'ণ', 'ত', 'থ', 'দ', 'ধ', 'ন', 'প', 'ফ', 'ব', 'ভ', 'ম', 'য', 'র', 'ল', 'শ', 'ষ',
    'স', 'হ', 'ড়', 'ঢ়', 'য়', 'ৎ', 'ঃ', 'ং', 'ঁ', '০', '১', '২', '৩', '৪', '৫', '৬', '৭', '৮', '৯', 'া', 'ি', 'ী',
    'ে', 'ু', 'faka', '্র', '্য', 'ক্ষ', 'ন্ত', 'ত্র', 'ঙ্গ', 'স্থ', 'স্ব', 'ক্ত', 'স্ত', 'ন্দ', 'চ্ছ', 'দ্ধ',
    'ন্ত্র', 'ফাকা', 'ত্ত', 'ষ্ট', 'ন্ন', 'ল্প', 'ম্প', 'faka', 'ূ', 'ৃ', 'ৈ', 'faka', 'ৌ', '।'
]

# Function to calculate accuracy for each image
def calculate_accuracy(predictions, ground_truths):
    correct_predictions = 0
    total_predictions = 0

    # Check if the predicted class matches any ground truth class
    for pred in predictions:
        if pred in ground_truths:
            correct_predictions += 1
        total_predictions += 1
    
    # Calculate accuracy
    accuracy = (correct_predictions / total_predictions) * 100
    return accuracy

# Function to calculate Character Error Rate (CER)
def calculate_cer(predictions, ground_truths):
    # CER is the ratio of incorrect predictions to the total characters in the ground truth
    incorrect_chars = 0
    total_chars = len(ground_truths)
    
    # Compare predicted classes to ground truth classes
    for pred, gt in zip(predictions, ground_truths):
        if pred != gt:
            incorrect_chars += 1
    
    # CER = incorrect characters / total characters in ground truth
    cer = incorrect_chars / total_chars
    return cer

# Lists to store all predictions and ground truths
all_predictions = []
all_ground_truths = []

# List for calculating total accuracy
all_accuracy = []
all_cer = []

# List all image files in the test directory
image_files = [f for f in os.listdir(test_images_dir) if f.endswith('.bmp')]

# Randomly select 10 images from the test set
random_images = random.sample(image_files, 10)

# Loop through the randomly selected 10 test images
for image_filename in random_images:
    # Construct full image and annotation paths
    image_path = os.path.join(test_images_dir, image_filename)
    annotation_path = os.path.join(test_annotations_dir, image_filename.replace('.bmp', '.txt'))

    # Load the ground truth classes from the annotation file (handle multiple classes)
    ground_truth_classes = []
    if os.path.exists(annotation_path):
        with open(annotation_path, 'r') as f:
            for line in f.readlines():
                parts = line.strip().split()
                class_id = int(parts[0])  # Ground truth class ID
                ground_truth_classes.append(class_id)

    # Ensure that there are ground truth classes before processing
    if len(ground_truth_classes) == 0:
        print(f"Warning: No ground truth for {image_filename}")
        continue  # Skip this image if there is no ground truth

    # Make predictions on the image
    results = model.predict(image_path, imgsz=640)  # Adjust imgsz if needed

    # Extract predicted class IDs from the results
    predicted_classes = results[0].boxes.cls.tolist()  # Get predicted class IDs

    # Ensure there are predictions
    if len(predicted_classes) == 0:
        print(f"Warning: No predictions for {image_filename}")
        continue  # Skip this image if no predictions

    # Calculate accuracy for this single prediction
    accuracy = calculate_accuracy(predicted_classes, ground_truth_classes)

    # Calculate CER for this single prediction
    cer = calculate_cer(predicted_classes, ground_truth_classes)

    # Store accuracy and CER for final calculation
    all_accuracy.append(accuracy)
    all_cer.append(cer)

    # Get the predicted class names for visualization
    predicted_class_names = [class_names[int(cls)] for cls in predicted_classes]
    ground_truth_class_names = [class_names[cls] for cls in ground_truth_classes]

    # Print the predicted classes, ground truth, and accuracy for each image
    print(f"Image: {image_filename}")
    print(f"Predicted Classes: {predicted_class_names}")
    print(f"Ground Truth Classes: {ground_truth_class_names}")
    print(f"Accuracy: {accuracy:.2f}%")
    print(f"CER: {cer:.2f}")
    print("-" * 50)

# Calculate the total accuracy and average CER
total_accuracy = sum(all_accuracy) / len(all_accuracy)
average_cer = sum(all_cer) / len(all_cer)

print(f"Overall Accuracy for 10 images: {total_accuracy:.2f}%")
print(f"Average CER for 10 images: {average_cer:.2f}")
```
